```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your datasets with the correct delimiter
math_df = pd.read_csv('student-mat.csv', delimiter=';')
por_df = pd.read_csv('student-por.csv', delimiter=';')

# Concatenate datasets
df = pd.concat([math_df, por_df], ignore_index=True)

# Print column names to verify
print("Column names in the dataset:")
print(df.columns)

# Strip leading/trailing spaces from column names
df.columns = df.columns.str.strip()

# List of attributes to plot frequencies for
attributes = ['studytime', 'failures', 'schoolsup', 'paid', 'higher', 'internet', 'health', 'absences']

# Plot frequency distributions
for attribute in attributes:
    if attribute in df.columns:
        plt.figure(figsize=(10, 6))
        
        if df[attribute].dtype == 'object':
            # Categorical attribute
            df[attribute].value_counts().plot(kind='bar', color='skyblue')
            plt.title(f'Frequency Distribution of {attribute}')
            plt.xlabel(attribute)
            plt.ylabel('Frequency')
        else:
            # Numerical attribute
            df[attribute].plot(kind='hist', bins=30, color='skyblue', edgecolor='black')
            plt.title(f'Frequency Distribution of {attribute}')
            plt.xlabel(attribute)
            plt.ylabel('Frequency')
        
        plt.grid(True)
        plt.show()
    else:
        print(f"Attribute '{attribute}' not found in the dataset.")
```
